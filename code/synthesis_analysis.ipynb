```python
import scripts.init_gpu as init_gpu
import scripts.init_dataset as init_dataset
from scripts.triplet_functions import n_neurons
import pandas as pd


init_gpu.initialize_gpus()

locations = ['LOC1', 'LOC2']

print("Loading Dataset...")
# load the dataset
df = pd.read_csv(
    f"../dataset/processed/{locations[0]}-{locations[1]}-scaled-balanced.csv")

length = len(df.columns) - 2  # subtract the two label columns

# limit the length
# length = 20
# start_idx, end_idx = 20, 40
# df = df.loc[:, ['Location', 'Website', *
#                     [str(i) for i in range(start_idx, end_idx)]]]
# get train-test set
train_df, test_df, train_web_samples, test_web_samples = init_dataset.get_sample(
    df, locations, range(1500), 1200)

train_df.head()
```

```python
# dataset for the classification
target_location = 'LOC2'
target_df = test_df[test_df['Location'] == target_location]
target_df.drop('Location', axis=1, inplace=True)
target_df.sort_values(by=['Website'], inplace=True)
target_df.reset_index(drop=True, inplace=True)
target_df.head(20)


train_location = 'LOC1'
train_location_df = test_df[test_df['Location'] == train_location]
train_location_df.drop('Location', axis=1, inplace=True)
train_location_df.sort_values(by=['Website'], inplace=True)
train_location_df.reset_index(drop=True, inplace=True)

train_location_df.info()
```

```python
synthetic_df = pd.read_csv('../synthesized/sampling-LOC1-target-LOC2.csv').iloc[:, 1:]
synthetic_df = synthetic_df[synthetic_df['Location'] == target_location]
synthetic_df.drop('Location', axis=1, inplace=True)
synthetic_df.sort_values(by=['Website'], inplace=True)
synthetic_df.reset_index(drop=True, inplace=True)
synthetic_df.head(20)
```

```python
def get_aggr_stats(df):
    gp = df.groupby(['Website'])
    # LOC1, LOC2 216 minimum number of samples
    # LOC3 some have only 99 samples

    return gp.mean(), gp.median(), gp.std()
```

```python
target_mean, target_median, target_std = get_aggr_stats(target_df)
synthetic_mean, synthetic_median, synthetic_std = get_aggr_stats(synthetic_df)
train_location_mean, train_location_median, train_location_std = get_aggr_stats(train_location_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def get_euclidean_distance(a, b):
    return np.sqrt(np.sum(np.square(a - b)))

def plot_mean_and_std(mean, std, website, label):
    plt.plot(mean.loc[website], label=f'{label} Mean')
    plt.fill_between(range(length), mean.loc[website] - std.loc[website], mean.loc[website] + std.loc[website], label=f'{label} Std', alpha = 0.1)

for website in test_web_samples:
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 2, 1)
    plot_mean_and_std(target_mean, target_std, website, f'Target {target_location}')
    plot_mean_and_std(synthetic_mean, synthetic_std, website, 'Synthetic')
    synth_eucld_dist = get_euclidean_distance(target_mean.loc[website], synthetic_mean.loc[website])
    plt.xticks(range(0,length, 10), labels=range(0, length, 10))
    plt.title(f"Synth vs Target - Website: {website}, Eucld Dist: {synth_eucld_dist:.3f}")
    plt.legend()
    
    plt.subplot(1, 2, 2)
    
    plot_mean_and_std(target_mean, target_std, website, f'Target {target_location}')
    plot_mean_and_std(train_location_mean, train_location_std, website, f'Train {train_location}')
    train_eucld_dist = get_euclidean_distance(target_mean.loc[website], train_location_mean.loc[website])
    plt.xticks(range(0,length, 10), labels=range(0, length, 10))
    plt.legend()
    plt.title(f"Train vs Target - Website: {website}, Eucld Dist: {train_eucld_dist:.3f}")
    plt.tight_layout()
    plt.savefig(f"../figures/Conditional-Synthesis-LOC2/{website}.png")
    plt.close()
    
```
