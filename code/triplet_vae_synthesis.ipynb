```python
import scripts.init_gpu as init_gpu
import scripts.init_dataset as init_dataset
from scripts.triplet_functions import n_neurons
import pandas as pd
from scripts.triplet_vae_model import VAE_Triplet_Model_V2, Sampling

init_gpu.initialize_gpus()

locations = ['LOC1', 'LOC2']

print("Loading Dataset...")
# load the dataset
df = pd.read_csv(
    f"../dataset/processed/{locations[0]}-{locations[1]}-scaled-balanced.csv")

length = len(df.columns) - 2  # subtract the two label columns

# get train-test set
train_df, test_df, train_web_samples, test_web_samples = init_dataset.get_sample(
    df, locations, range(1500), 1200)

train_df.head()
```

```python
import tensorflow as tf
@tf.function
def synthesize_traffic(model, content, source_location, target_location):
    content_tensor = tf.convert_to_tensor(content, dtype=tf.float32)
    z_mean, z_log_var, z = model.encode(content_tensor)
    
    target_location_tensor = tf.convert_to_tensor(target_location, dtype=tf.float32)
    target_location_tensor = tf.reshape(target_location_tensor, (1, -1))
    target_location_tensor = tf.tile(target_location_tensor, [tf.shape(z)[0], 1])
    
    # Decode using the latent representation and target location
    synthesized = model.decode(z, target_location_tensor)
    return synthesized
```

```python
import numpy as np
def location_to_onehot(select_location, locations):
    if select_location not in locations:
        raise ValueError(f"'{select_location}' is not in the list of locations")
    
    return np.array([1 if location == select_location else 0 for location in locations]).astype(np.float32)
location_to_onehot_dict = {location: location_to_onehot(location, locations) for location in locations}
location_to_onehot_dict
```

```python
model = tf.keras.models.load_model('../models/triplet-vae/triplet_vae-v2-e150-mse1000-kl1-t100-LSTM-CD10.keras', custom_objects={'VAE_Triplet_Model_V2': VAE_Triplet_Model_V2, 'Sampling': Sampling})
```

```python
def filter_and_sort_data(df, location_label):
    """
    Filter the dataframe by location label, sort by Website label, and drop specified columns.
    """
    return (df[df['Location'] == location_label]
            .sort_values(by=['Website'])
            .iloc[:, 2:]
            .reset_index(drop=True))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from fastdtw import fastdtw

import numpy as np
def get_euclidean_distance(a, b):
    return np.sqrt(np.sum(np.square(a - b)))


def evaluate_decoder(sampling_location, target_location):
    synthesized = synthesize_traffic(
    model, 
    filter_and_sort_data(test_df, sampling_location).to_numpy().astype(np.float32),  # Convert to float32 here
    sampling_location,
    location_to_onehot_dict[target_location]
    )
    target_traces = filter_and_sort_data(test_df, target_location)

    eucld_dists = []
    
    num_synth_websites = 20
    for i in range(0, 200 * num_synth_websites, 200):
        plt.figure()
        # real mean
        real_mean = np.mean(target_traces[i:i+200], axis=0)
        real_std = np.std(target_traces[i:i+200])
        
        plt.plot(real_mean, label='real')
        plt.fill_between(range(length),real_mean + real_std, real_mean - real_std, label='real std', alpha = 0.2)
        
        # synth mean
        synthetic_mean = np.mean(synthesized[i:i+200], axis=0)
        synthetic_std = np.std(synthesized[i:i+200])
        
        dtw_distance, path = fastdtw(real_mean, synthetic_mean, dist=get_euclidean_distance)
        euclidean_distance = get_euclidean_distance(real_mean, synthetic_mean)

        plt.plot(synthetic_mean, label='synthetic mean')
        plt.fill_between(range(length), synthetic_mean + synthetic_std, synthetic_mean - synthetic_std, label='synth. std', alpha = 0.2)
        
        
        plt.xticks(range(0,length, 10), labels=range(0, length, 10))
        plt.title(f"{target_location}, {test_web_samples[i // 200]}, DTW: {dtw_distance:.2}, Ecld_dist: {euclidean_distance:.2}")
        plt.legend()
        plt.show()
        eucld_dists.append(euclidean_distance)
        

    print(sum(eucld_dists) / num_synth_websites)
```

```python
evaluate_decoder('LOC1', 'LOC2')
```
